```python
import numpy as np
import matplotlib.pyplot as plt
def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = [(data - mean) / std_dev for x in data ]
    return z_scores
data = np.random.normal(loc=0, scale=1, size=1000)
z_scores = calculate_z_scores(data)
plt.plot(z_scores, marker='o', linestyle='')
plt.xlabel('Data Points Index')
plt.ylabel('Z-Scores')
plt.title('Z-Score Plot')
plt.show()
```

19. Implement hypothesis testing using Z-statistics for a dataset.

```python
import numpy as np
from scipy import stats
def z_test(sample_means, population_mean, population_std_dev, sample_size):
    z_score = (sample_means - population_mean) / (population_std_dev / np.sqrt(sample_size))
    p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))
    return z_score, p_value
sample_means = 50
population_mean = 50
population_std_dev = 5
sample_size = 100
z_score, p_value = z_test(sample_means, population_mean, population_std_dev, sample_size)
print("Z-score:", z_score)
print("P-value:", p_value)
```

20. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats
def confidence_interval(data, confidence_level=0.95):
    a=1.0 * np.array(data)
    n = len(a)
    mean, std_dev = np.mean(a), np.std(a)
    h = std_dev * stats.t.ppf((1 + confidence_level) / 2., n-1)
    start = mean - h
    end = mean + h
    return start, end
data=np.random.normal(0,1,100)
lower, upper = confidence_interval(data)
print("Confidence Interval:", (lower, upper))

```

28.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis
def calculate_skewness_kurtosis(size=1000, mean=0, std_dev=1):
    dataset = np.random.normal(mean, std_dev, size)
    skewness = skew(dataset)
    kurtosis_value = kurtosis(dataset)
    return skewness, kurtosis_value
skewness,kurtosis_value=calculate_skewness_kurtosis()
print("Skewness:", skewness)
print("Kurtosis:", kurtosis_value)

```
