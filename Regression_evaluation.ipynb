# REGRESSION AND ITS EVALUATION

1. What is Simple Linear Regression ?

--> Simple Linear Regression is a statistical method used to model th relationship between two continous variables. It involves creating a linear equation that best predicts the value of one variable based on the value of another variable.

2. What are the key assumptions of Simple Linear Regression ?

--> The key assumption of Simple Linear Assumption.


1.   Linearity :The relationship between the independent and dependent          variables should be linear .

2.   Independences :observation should be independent of each other.

3. Homoscedasticity : The variance of the residuals should be constant across all levels of the independent variable.

4. Normality : The residuals should be normaly distributed.

5. No multicollinearity : The independent variables should not be highly correlated with each other.









3. What is heteroscedasticity , and why is it important to address in regression model's ?

--> Heteroscedasticity is the condition of non-constant variance of residuals , and it's cruical to address it to ensure accurate and reliable regression analysis,

4. What is Multiple Linear REgression ?

--> Multiple Linear Regression is an extension of Simple Linear Regression that involves more than one independendent variable to predict the outcome of a dependent variable . It helps in understanding how multiple factors influence a particular outcomes.

5. What is polynomial regression , and how does it differ from linear regression ?

--> Polynomial regression is a form of regrssion analysis where the realtionship between the independent variable and the dependent variable is modeled as an nth degree polynomial.

It differs from linear regression , which models the relationship as a straight line(first-degree polynomial) by allowing for more complex , curved realtionships through the use of higher-degree polynomials.

6. Implement a Python program to fit a Simple Linear Regresssion model to the following sample data #


*   X=[1,2,3,4,5]
*   Y=[2.1, 4.3,6.1, 7.9, 10.2 ]

plot the regression over the data points.





```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([2.1, 4.3,6.1, 7.9, 10.2])
model = LinearRegression()
model.fit(X,Y)
Y_pred = model.predict(X)
plt.scatter(X,Y, color='green', label='Data Points')
plt.plot(X,Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.plot(X,Y_pred,color='red')
plt.show()
print(f"Intercept : {model.intercept_}")
print(f"Slope : {model.coef_[0]}")
```

7. Fit a Multiple Linear Regression model on this sample data:


*   Area =  [ 1200, 1500, 1800, 2000 ]
*   Rooms = [2,3,3,4]
*   Price = [250000, 300000, 320000, 370000 ]

Check for multicollinearity using VIF and report the results.



```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
Area = np.array([1200, 1500, 1800, 2000])
Rooms = np.array([2,3,3,4])
Price = np.array([250000, 300000, 320000, 370000])
df = pd.DataFrame({'Area':Area, 'Rooms':Rooms, 'Price':Price})
X = df[['Area', 'Rooms']]
Y = df['Price']
model = LinearRegression()
model.fit(X,Y)
Y_pred = model.predict(X)
model = LinearRegression()
model.fit(X,Y)
Y_pred = model.predict(X)
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif_data)
print(f"Intercept : {model.intercept_}")
print(f"Slope : {model.coef_}")
```

8, Implement polynomial regression on the following data ⁉



*   X = [1,2,3,4,5]
*   Y = [ 2.2, 4.8, 7.5, 11.2, 14.7 ]

Fit a 2nd degree polynomial and plot the resulting curve.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])
degree = 2
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)
model = LinearRegression()
model.fit(X_poly,Y)
X_test = np.linspace(0,5,100).reshape(-1,1)
X_test_poly = poly_features.transform(X_test)
Y_pred = model.predict(X_test_poly)
plt.scatter(X,Y, color='green', label='Data Points')
plt.plot(X_test,Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

9. Create a residuals plot for a regression model trained on this data :

*  X = [10, 20, 30, 40, 50]  
*  Y = [ 15,35, 40, 50,65]

Asses heteroscedasticity by examining the spread of residuals.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X = np.array([10, 20, 30, 40, 50]).reshape(-1,1)
Y = np.array([15, 35, 40, 50,65])
model = LinearRegression()
model.fit(X,Y)
Y_pred = model.predict(X)
residuals = Y - Y_pred
plt.scatter(Y_pred, residuals, color='green')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Prdicted Values')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.show()
```
